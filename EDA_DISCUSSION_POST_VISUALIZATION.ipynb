```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = {
    "Date": ["2005-02-25", "2005-02-28", "2005-03-01", "2005-03-02", "2005-03-03", "2005-03-04"],
    "Open": [46.177, 46.525, 46.481, 46.369, 46.571, 46.772],
    "High": [46.562, 46.525, 46.543, 46.735, 46.597, 47.035],
    "Low": [46.177, 46.186, 46.43, 46.369, 46.333, 46.772],
    "Close": [46.534, 46.289, 46.543, 46.571, 46.589, 46.999],
    "Volume": [1484, 4341, 2628, 393130, 13705, 16792],
    "OpenInt": [0, 0, 0, 0, 0, 0]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])
df["Volatility"] = df["High"] - df["Low"]
df["Price Movement"] = df["Close"] - df["Open"]

# Plot time series
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], label="Closing Price", color="blue")
plt.bar(df["Date"], df["Volume"], label="Volume", color="orange", alpha=0.5)
plt.title("Closing Prices and Volumes Over Time")
plt.xlabel("Date")
plt.ylabel("Price / Volume")
plt.legend()
plt.show()

# Scatter plot for volume vs price movement
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Volume"], y=df["Price Movement"])
plt.title("Volume vs. Price Movement")
plt.xlabel("Volume")
plt.ylabel("Price Movement")
plt.show()
```
